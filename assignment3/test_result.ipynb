```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt

from utils import ackley_f, weierstrass_f
from CSO_algorithm import cuckoo_search
from BAT_algorithm import bat_algorithm
```

# 1. Cuckoo Search

## 1.1. Ackley function

```python
cso_ackley_10dim = cuckoo_search(function=ackley_f)
cso_ackley_20dim = cuckoo_search(function=ackley_f, dimension=20)
```

```python
all_values_list_cso_ackley_10d = []
all_values_list_cso_ackley_20d = []
```

```python
for i in range(20):
    values, points = cso_ackley_10dim.forward(350)
    all_values_list_cso_ackley_10d.append(array(values))
```

```python
for i in range(20):
    values, points = cso_ackley_20dim.forward(500)
    all_values_list_cso_ackley_20d.append(array(values))
```

```python
all_values_matrix_cso_ackley_10d = np.row_stack(all_values_list_cso_ackley_10d) # array(number of test, number of iteration)

max_line_cso_ackley_10d = np.max(all_values_matrix_cso_ackley_10d, axis=0)
mean_line_cso_ackley_10d = np.mean(all_values_matrix_cso_ackley_10d, axis=0)
min_line_cso_ackley_10d = np.min(all_values_matrix_cso_ackley_10d, axis=0)

all_values_matrix_cso_ackley_20d = np.row_stack(all_values_list_cso_ackley_20d) # array(number of test, number of iteration)

max_line_cso_ackley_20d = np.max(all_values_matrix_cso_ackley_20d, axis=0)
mean_line_cso_ackley_20d = np.mean(all_values_matrix_cso_ackley_20d, axis=0)
min_line_cso_ackley_20d = np.min(all_values_matrix_cso_ackley_20d, axis=0)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(range(351), mean_line_cso_ackley_10d)
plt.fill_between(range(351), min_line_cso_ackley_10d, max_line_cso_ackley_10d, color="b", alpha=0.1, label="dimension-10")

plt.plot(range(501), mean_line_cso_ackley_20d)
plt.fill_between(range(501), min_line_cso_ackley_20d, max_line_cso_ackley_20d, color="r", alpha=0.1, label="dimension-20")
plt.xlabel("Number of iterations", weight="bold", size=12)
plt.ylabel("function values", weight="bold", size=12)
plt.title("Ackley function", weight="bold", size=15)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
